# ***Playing a Video With OS***
When a video file is opened and played on a computer, the operating system uses the file system to locate and open the file. The file is then read from the hard disk or storage device and copied into the computer's RAM. The CPU then uses a codec to decode the video and audio data, which is then sent to the graphics card for rendering and display on the monitor. The audio data is sent to the audio card for playback through the speakers or headphones. The CPU also coordinates the different processes involved in playing the video, such as adjusting the volume, seeking to different parts of the video, and pausing or stopping playback.


# ***Open a huge Text File using OS***
When a file is opened, it is loaded into the computer's RAM (Random Access Memory) for faster access. The RAM is a temporary storage area that the computer uses to access data quickly. When a file is opened, the computer reads the entire file into RAM, which is why the file takes up more space in RAM than it does on a disk drive. Additionally, the operating system and any other programs that are running on the computer also take up space in RAM, which can cause the file to take up even more space.
